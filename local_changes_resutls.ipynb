# Results for Local Changes

```python
from pathlib import Path
import json


local_changes_path = Path("Local Changes")

results_files = list(local_changes_path.rglob("*_results.json"))
results_files

# Find all files with *_results.json
# Get the cubicles names

dic_of_cubicles = {}
for file in results_files:
    with open(file, "r") as f:
        data = json.load(f)
    dic_of_cubicles[file.parts[-1].split("_")[0]] = data


dic_of_cubicles
```

```python
import matplotlib.pyplot as plt

names_of_cubicles = list(dic_of_cubicles.keys())
scores_of_cubicles = [dic_of_cubicles[name]["score"]["percentage"] for name in names_of_cubicles]


plt.bar(names_of_cubicles, scores_of_cubicles)
plt.xlabel("Cubicles")
plt.ylabel("Accuracy [%]")
plt.title("Scores of each cubicle")
plt.xticks(rotation=45)
plt.show()
# Get the scores
# Display in a bar chart the scores of each cubicle
```
